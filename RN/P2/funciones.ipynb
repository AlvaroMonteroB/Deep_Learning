```python
import numpy as np
import matplotlib.pyplot as plt

#Escalon
x=(-10,0,10)
y=(0,0,1)

plt.step(x,y,where='pre',color='red')
plt.plot(x,y,'C0o',alpha=.5)
plt.show
```

```python
#Lineal
x=np.linspace(-5,5,100)
y=2*x
plt.plot(x,y,'-r',label='Y=2x',color='blue')
plt.title("Graph of the linear function")
plt.xlabel('x',color='blue')
plt.ylabel('y',color='blue')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
#Sigmoide
def logistic(x):
    return (1/(1+np.exp(-x)))
plt.title("Graph of logistic function")
plt.plot(x,logistic(x),color='black')
plt.grid()
plt.show()
```

```python
#Tangente hiperbolica
def tanh(x):
    return (2/(1+np.exp(-2*x))-1)
plt.title("Graph of logistic function")
plt.plot(x,tanh(x),color='black')
plt.grid()
plt.show()
```

```python
#Relu
def relu(x):
    y=np.zeros(x.shape)
    return np.ndarray.flatten(np.max([[y],[x]],axis=0))
plt.title("Graph of relu function")
plt.plot(x,relu(x),color='black')
plt.grid()
plt.show()
```

```python
#Softmax
import scipy.special

def softmax_grad(s):
    jacobian=np.diag(s)
    
    for i in range(len(jacobian)):
        for j in range(len(jacobian)):
            if i==j:
                jacobian[i][j]=s[i]*(1-s[i])
            else:
                jacobian[i][j] = -s[i] * s[j]
    return jacobian

x=np.array([1,2])
scipy.special.softmax(x)
softmax_grad(scipy.special.softmax(x))
    
```
