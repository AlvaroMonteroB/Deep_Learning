```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

celsius=np.array([-40,-10,0,8,15,22,38,42,30],dtype=float)
farenheit=np.array([-40,14,32,46,59,72,100,107.6,86])
```

Capas y modelo

```python
capa=tf.keras.layers.Dense(units=1,input_shape=[1])
modelo=tf.keras.Sequential([capa])

modelo.compile(
    optimizer=tf.keras.optimizers.Adam(.1),
    loss="mean_squared_error"
)
```

Entrenamiento

```python
historial=modelo.fit(celsius,farenheit,epochs=1000,verbose=False)
plt.xlabel("# epoca")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history["loss"])
plt.show()
```

Predicci√≥n

```python
result=modelo.predict([100.00])
print("El resultado es "+str(result)+" Farenheit")

print("Variables internas del modelo")
print(capa.get_weights())
```
