```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
millas=np.array([0,1,3,5,7,8.2,9.1,10.4,.33,11.7,13.7])
km=np.array([0,1.60934,4.82803,8.04672,11.2654,13.1966,14.645,16.73718,0.531084,18.82932,22.04801])
modelo=tf.keras.Sequential(
   [ tf.keras.layers.Dense(units=3,input_shape=[1]),
    tf.keras.layers.Dense(units=3),
    tf.keras.layers.Dense(units=1)]
)
modelo.compile(
    optimizer="adam",
    loss="mean_squared_error"
)
```

```python
plt.plot(millas, km, marker='o')  # graficar millas en el eje x y km en el eje y
plt.xlabel('Millas')  # etiqueta del eje x
plt.ylabel('KiloMetros')  # etiqueta del eje y
plt.title('Datos')  # título del gráfico
plt.grid(True)  # activar la cuadrícula en el gráfico
plt.show()  # mostrar el gráfico
```

```python
history=modelo.fit(millas,km,epochs=2000)
plt.xlabel("# epoca")
plt.ylabel("Magnitud de perdida")
plt.plot(history.history["loss"])
plt.show()
```

```python
x=np.linspace(0,20,20)
test_values=modelo.predict(x)
plt.plot(x, test_values, marker='o')  # graficar millas en el eje x y km en el eje y
plt.xlabel('Millas')  # etiqueta del eje x
plt.ylabel('Metros')  # etiqueta del eje y
plt.title('Conversión de Millas a Metros')  # título del gráfico
plt.grid(True)  # activar la cuadrícula en el gráfico
plt.show()  # mostrar el gráfico
```

```python
lista=[(milla,kilometro) for milla,kilometro in zip(x,test_values)]
print(lista)
```

```python
modelo.summary()
w=modelo.get_weights()
print(w)
print("\nObjeto Pesos:" , w)
print("Number of weights "+str(len(w)))
print("\nw= "+ str(w[0][0])+"(Weight)")
print("b = "+str(w[1])+" Weight->bias")
```

```python

x=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,0,0,1])

and_gate=tf.keras.Sequential([
    tf.keras.layers.Dense(units=1,input_shape=[2],activation="sigmoid")
])
and_gate.compile(
    optimizer="adam",
    loss='mean_squared_error',
    metrics=["accuracy"]
)
```

```python
history=and_gate.fit(x,y,epochs=1000)
plt.xlabel("# epoca")
plt.ylabel("Magnitud de perdida")
plt.plot(history.history["loss"])
plt.show()
```

```python
pred=and_gate.predict(np.array([[0,1]]))
print(pred)
```

```python
w=and_gate.get_weights()
print(w)
print("\nObjeto Pesos:" , w)
print("Number of weights "+str(len(w)))
print("\nw= "+ str(w[0][0])+"(Weight)")
print("b = "+str(w[1])+" Weight->bias")
```
